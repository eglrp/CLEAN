```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
with open("origin.txt") as fin:
    x,y = [], []
    for line in fin.readlines():
        x.append(int(line.split(" ")[0]))
        y.append(float(line.split(" ")[1]))
with open("cprs_exp_DP.txt") as fin:
    cx,cy = [], []
    for line in fin.readlines():
        cx.append(int(line.split(" ")[0]))
        cy.append(float(line.split(" ")[1]))
```

```python
x =[i - x[0]for i in x]
y = [i - y[0] for i in y]
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(16,12))
plt.plot(x, y, marker= '*',label='Origin T&D')
plt.plot(cx, cy, marker= '^',label='EXP_DP T&D span=3')
plt.legend()
```
